```python
import pandas as pd
import numpy as np
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Book4/Ch10/Bike_distance2.xlsx')
df.shape
```

```python
df.head(3)
```

```python
data = df.drop(['Rented Bike B'], axis=1)   # 타깃 변수를 제외한 변수만 data에 저장
target = df['Rented Bike B']                # 타깃 변수만 target에 저장
```

```python
# 50:50 데이터 분할
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    data, target, test_size=0.5, random_state=42)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
# Logistic Regression 모형 (Default 모델)
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr = LogisticRegression(penalty='none',random_state=42)
model = lr.fit(X_train, y_train)
pred = model.predict(X_test)

print ("Logreg Training set score:{:.5f}".format(model.score(X_train, y_train)))
print ("Logreg Test set score:{:.5f}".format(accuracy_score(y_test, pred)))
```

```python
# Logistic Regression 릿지 모형 (L2 제약)
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr = LogisticRegression(penalty='l2',random_state=42)
model = lr.fit(X_train, y_train)
pred = model.predict(X_test)

print ("Logreg Training set score:{:.5f}".format(model.score(X_train, y_train)))
print ("Logreg Test set score:{:.5f}".format(accuracy_score(y_test, pred)))
```

```python
# Logistic Regression 라쏘 모형 (L1 제약)
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr = LogisticRegression(penalty='l1',solver='liblinear', random_state=42)
model = lr.fit(X_train, y_train)
pred = model.predict(X_test)

print ("Logreg Training set score:{:.5f}".format(model.score(X_train, y_train)))
print ("Logreg Test set score:{:.5f}".format(accuracy_score(y_test, pred)))
```

```python
# Logistic Regression 모형 (Default 모델)
lr = LogisticRegression(solver='lbfgs',penalty='none',random_state=42)
```

```python
# 그리드 서치 Logistic Regression 모델
from sklearn.model_selection import GridSearchCV
# solver에 liblinear는 none 제약과 맞지 않아 제거
# 타 solver값을 추가하면 최종 성능이 떨어져서 제거
params = {'solver':['lbfgs','sag','saga'],\
          'penalty':['none'], 'max_iter':[100, 1000, 10000]}

grid_lr = GridSearchCV(lr, param_grid=params, scoring='accuracy', cv=5, n_jobs=-1)
grid_lr.fit(X_train, y_train)

print("GridSearchCV max accuracy:{:.5f}".format(grid_lr.best_score_))
print("GridSearchCV best parameter:", (grid_lr.best_params_))
```

```python
best_clf = grid_lr.best_estimator_
pred = best_clf.predict(X_test)
print("Accuracy on test set:{:.5f}".format(accuracy_score(y_test, pred)))
```

```python
# Logistic Regression 릿지 모델 (Default 모델)
lr = LogisticRegression(penalty='l2',random_state=42)
```

```python
# 그리드 서치
from sklearn.model_selection import GridSearchCV
# 아래 solver에 추가 방법이 하나 더 있으나, 이를 추가하면 최종 성능이 떨어져 넣지 않음
params = {'solver':['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],\
          'penalty':['l2'], 'max_iter':[100, 1000, 10000]}

grid_lr = GridSearchCV(lr, param_grid=params, scoring='accuracy', cv=5, n_jobs=-1,)
grid_lr.fit(X_train, y_train)

print("GridSearchCV max accuracy:{:.5f}".format(grid_lr.best_score_))
print("GridSearchCV best parameter:", (grid_lr.best_params_))
```

```python
best_clf = grid_lr.best_estimator_
pred = best_clf.predict(X_test)
print("Accuracy on test set:{:.5f}".format(accuracy_score(y_test, pred)))
```

```python
# Logistic Regression 라쏘 모델 (Default 모델)
lr = LogisticRegression(penalty='l1',random_state=42)
```

```python
# 그리드 서치
from sklearn.model_selection import GridSearchCV
# 아래 두 solver값만 L1 페널티에 해당
params = {'solver':['liblinear','saga'],\
          'penalty':['l1'], 'max_iter':[100, 1000, 10000]}

grid_lr = GridSearchCV(lr, param_grid=params, scoring='accuracy', cv=5, n_jobs=-1,)
grid_lr.fit(X_train, y_train)

print("GridSearchCV max accuracy:{:.5f}".format(grid_lr.best_score_))
print("GridSearchCV best parameter:", (grid_lr.best_params_))
```

```python
best_clf = grid_lr.best_estimator_
pred = best_clf.predict(X_test)
print("Accuracy on test set:{:.5f}".format(accuracy_score(y_test, pred)))
```

# SVM

```python
# SVM model (default 모델)

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svm = SVC(kernel='rbf', C=1, random_state=42, probability=True)
                                                     # probability=True 에 주의
model = svm.fit(X_train, y_train)
pred = model.predict(X_test)  # 학습된 Classifier로 테스트 데이터셋 자료이용해서 타겟변수 예측값 생성
accuracy = accuracy_score(y_test, pred)

print ("SVM Accuracy on training set:{:.5f}".format(model.score(X_train, y_train)))
print ("SVM Accuracy on test set:{:.5f}".format(accuracy_score(y_test, pred)))
```

```python
# SVM model 2

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svm = SVC(kernel='rbf', C=1, random_state=42, gamma='scale', probability=True)
                                                     # probability=True 에 주의
model = svm.fit(X_train, y_train)
pred = model.predict(X_test)  # 학습된 Classifier로 테스트 데이터셋 자료이용해서 타겟변수 예측값 생성
accuracy = accuracy_score(y_test, pred)

print ("SVM Accuracy on training set:{:.5f}".format(model.score(X_train, y_train)))
print ("SVM Accuracy on test set:{:.5f}".format(accuracy_score(y_test, pred)))
```

```python
# 그리드 서치
# SVM model (default 모델)
svm = SVC(kernel='rbf', C=1, random_state=42, probability=True)
                                                     # probability=True 에 주의

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold

# StratifiedKFold의 random_state 옵션값을 특정 숫자(예: 0)로 고정
cross_validation = StratifiedKFold(n_splits=3, shuffle=True, random_state=0)
params = {'kernel':['linear', 'rbf', 'sigmoid'], 'C':[1],
          'gamma':['scale']}

grid_svm = GridSearchCV(svm, param_grid=params, scoring='accuracy',
                        cv=cross_validation, n_jobs=-1)
grid_svm.fit(X_train, y_train)

print("GridSearchCV max accuracy:{:.5f}".format(grid_svm.best_score_))
print("GridSearchCV best parameter:", (grid_svm.best_params_))
```

```python
best_clf = grid_svm.best_estimator_
pred = best_clf.predict(X_test)
print("Accuracy on test set:{:.5f}".format(accuracy_score(y_test, pred)))
```
