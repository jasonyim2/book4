# 연속변수 타겟일 경우 회귀(Linear Regression) 모델

```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Book4/Ch9/Apart_distance1.csv')
df.shape
```

```python
df.drop(['거래금액'], axis=1, inplace=True) #'거래금액' 변수 제외
df.shape
```

```python
df.columns
```

```python
data = df.drop(['거래금액1'], axis=1)   # 타깃 변수를 제외한 변수만 data에 저장
target = df['거래금액1']                # 타깃 변수만 target에 저장
```

```python
# 80:20 data partition.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    data, target, test_size=0.2, random_state=42)   # test_size=0.2임에 주의

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
# 연속변수 타겟변수일 때 Linear Regression 모델 (Default 모델)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

linr = LinearRegression(n_jobs=-1)
model = linr.fit(X_train, y_train)
pred = model.predict(X_test)

print ("Linear Regression Training set r2 score:{:.5f}".format(model.score(X_train, y_train)))
print ("Linear Regression Test set r2 score:{:.5f}".format(r2_score(y_test, pred)))
```

# 연속변수 타겟일 경우 릿지(Ridge) 및 라쏘(Lasso) 모델

```python
# 연속변수 타겟변수일 때 Ridge 모델 (Default 모델)
from sklearn.linear_model import Ridge
Ridge = Ridge(random_state=42)
model = Ridge.fit(X_train, y_train)
pred = model.predict(X_test)

print ("Linear Regression Training set score:{:.5f}".format(model.score(X_train, y_train)))
print ("Linear Regression Test set score:{:.5f}".format(r2_score(y_test, pred)))
```

```python
# 연속변수 타겟변수일 때 Lasso 모델 (Default 모델)
from sklearn.linear_model import Lasso
Lasso = Lasso(random_state=42)
model = Lasso.fit(X_train, y_train)
pred = model.predict(X_test)

print ("Linear Regression Training set score:{:.5f}".format(model.score(X_train, y_train)))
print ("Linear Regression Test set score:{:.5f}".format(r2_score(y_test, pred)))
```

```python
# 책의 16개 변수 선택을 위한 Lasso 모델
from sklearn.linear_model import Lasso
Lasso = Lasso(random_state=42, alpha=0.01)
model = Lasso.fit(X_train, y_train)
pred = model.predict(X_test)

print ("Linear Regression Training set score:{:.5f}".format(model.score(X_train, y_train)))
print ("Linear Regression Test set score:{:.5f}".format(r2_score(y_test, pred)))
```
